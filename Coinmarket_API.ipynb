```python
# Create a grouped DataFrame with percentage changes across different timeframes
df3 = df.groupby('name', sort = False)[['quote.USD.percent_change_1h', 'quote.USD.percent_change_24h', 'quote.USD.percent_change_7d', 'quote.USD.percent_change_30d', 'quote.USD.percent_change_60d', 'quote.USD.percent_change_90d']].mean()
df3
```

```python
# Stack the DataFrame to prepare for visualization
df4 = df3.stack()
df4
```

```python
# Convert stacked Series to DataFrame
df5 = df4.to_frame(name = 'values')
df5
```

```python
# Rename columns for clarity
df7 = df6.rename(columns = {'level_1':'percent_change'})
df7
```

```python
# Clean up percent_change column by removing prefix and underscores
df7['percent_change'] = df7['percent_change'].str[-3:]
df7['percent_change'] = df7['percent_change'].str.replace('_', '', regex=False)
```
